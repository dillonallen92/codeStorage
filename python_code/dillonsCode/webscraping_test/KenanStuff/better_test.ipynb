```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
ids = []
initUrl = "https://soccer24.com/england/premier-league/results/"
statsUrl = "https://www.soccer24.com/match/
statsAdd = "/#match-statistics;0"
```

```python
driver.get(initUrl)
```

```python
vals = driver.find_elements_by_class_name("event__match--oneLine")
```

```python
for v in vals:
    print(v.text)
```

```python
vals2 = driver.find_elements_by_xpath("//div[starts-with(@id,'g_1_')]")
```

```python
for v in vals2:
    print(v.get_attribute('id'))
    ids.append(v.get_attribute('id'))
```

```python
for e in ids:
    print(e)
```

```python
def grabDataFromID(uID):
    targetUrl = statsUrl + uID + statsAdd
    driver.get(targetUrl)
```
