```python
# RQ1 Test
# i is the person, starting at 0 and ending at 358
rq1COUNT = 0
for i in range(df.shape[0]):

    q6Count  = 0
    q8Count  = 0
    q13Count = 0
    q16Count = 0
    q6_val   = df.iat[i,q6Idx]
    q8_val   = df.iat[i,q8Idx]
    q13_val  = df.iat[i,q13Idx]
    q16_val  = df.iat[i,q16Idx]

    if (q6_val == 'Yes'):
        q6Count += 1
    if ((q8_val == '4 - Slightly Affected') or (q8_val == '5 - Affected')):
        q8Count = q8Count + 1
    if((q13_val == '1 - Strongly Agree') or (q13_val == '2 - Agree')):
        q13Count = q13Count + 1
    if((q16_val == '4 - Slightly Affected') or (q16_val == '5 - Affected')):
        q16Count = q16Count + 1
    
    qCountSum = q6Count + q8Count + q13Count + q16Count

    if(qCountSum >= 2):
        rq1COUNT = rq1COUNT + 1
print("RQ1 Count: {}".format(rq1COUNT))
```
